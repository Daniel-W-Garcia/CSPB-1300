```python
#any time there are multiple values assigned to a single variable separated by commas python will make it a tuple

dg = 46, 1980, 'dad', 'tall'

# you do not need to wrap tuples in () when it's listed this way.
#eq of above:

dg1 = (46, 1980, 'dad', 'tall')

print(type(dg))
print(type(dg1))
```

```python
#unpacking tuples. You can do a single line to assign all values in a tuple to a variable:

#this is like writing individual variable assignments based on index of the tuple
age, year, status, height = dg

print(age, year, status, height)
```

```python
#using a tuple to swap variables
#example
a = 1
b = 2

#to swap we need a temp variable (like when we manually sort)
temp = a
a = b
b = temp
print(a,b, temp) # values swapped.

#with a tuple instead:
c = 3
d = 4
(c,d) = (d,c)
print(c,d)# values swapped without the need for a temp
```

```python
#unpacking into iterator variables:
players = [('Jordan', 'Michael'), ('Barkley', 'Charles'), ('Pippen', 'Scottie')]

#Assigning variable names as iterator var
for last_name, first_name in players: # every iteration unpacks the tuple into 2 vars
    print('First name:', first_name, 'Last name:', last_name)# flipping the order of values displayed
```

```python
dogs = ['lab', 'pit', 'gsd', 'mutt']

for n in range(len(dogs)):
    print(n, dogs[n])

print()
print('--------- Using enum-----------')

# keyword enumerate takes a sequence as input and returns a sequence of tuples
fruits = ['apple', 'pear', 'apricot', 'cherry', 'peach']

#enumerate provides an index for each item in the input sequence
for item in enumerate(fruits):
    print(item[0], item[1])
```

```python
#the 'pythonic' way is to unpack tuples while iterating:
drinks = ['water', 'beer', 'soda', 'wine']

for indx, drink in enumerate(drinks): # this is more succinct. assigns variable name for the index and for the item in sequence
    print(indx, drink)
```

```python
pokemon = {'Rattata': 19, 'Machop': 66, 'Seel': 86, 'Volbeat': 86, 'Solrock': 126}

p_names = []
p_number = []

#unpack kvp as tuples into individual var names (.items() returns a list of kvp tuples)
for name, num in pokemon.items():
    p_names.append(name)# append key
    p_number.append(num)# append value

print('Names:', p_names)
print('Numbers:', p_number)
```

```python
#unpack tuples to use in function calls:

def add(x,y): #requires 2 inputs
    return x + y
print(add(3,4))

z = (5,6) # this is a tuple with 2 values
print(add(*z)) #using '*' will unpack the tuple to use in the function

```
