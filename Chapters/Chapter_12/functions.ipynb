```python
def greeting(): # The 'def' keyword is required for all functions as well as '()' regardless of param required
    """ This line is knows as the 'docstring'. It is used as documentation and can be called with <function_name>._doc_"""
    print('Hello!')
    print('Glad you are here')
greeting() # calling the function with 0 params specified
```

```python
def add_2_nums(num1, num2): # function requires 2 params to work
    """ Adds 2 numbers together""" # this will print when add_2_nums.__doc__ is called
    return num1 + num2

print(add_2_nums(5,9))
print(add_2_nums(5 * 4, 5)) # you can place expressions as a param as long as it's the correct type. Params are separated by commas
add_2_nums.__doc__
```

```python
# Functions that return a value are sometimes referred to as 'Fruitful Functions'
# All functions in python return a value. If there is no specified 'return' then 'None' is returned
# a Function will end one it have returned a value. So if the function has several return statements, whenever the function
# gets to a place where it can return a value, none of the remaining code will execute

def multiply_two(int1, int2):
    """ Returns product of 2 given numbers"""
    y = int1 *int2
    return y
print(multiply_two(4, 5))

def multi_returns():
    print('First return:')
    return 10
    print('Second return:') #this is never reached since the func already executed a return.
    return 1000 # never reached
```

```python
names = ['dan', 'mike', 'jeff', 'angela']

def longer_than_five(list_of_names):
    """ Determines if any names in a given list are longer than 5 chars"""
    for name in names:
        if len(name) > 5:
            long_name = True
        else:
            long_name = False
    return long_name

print(longer_than_five(names))
```

```python
'''
Few things happening here. "message: str) is informing that param input type should be string, This is NOT enforced 
Then the -> annotation (optional) is telling what type of value is returned by the function
'''

def duplicate (message: str) -> str:
    return message+message

print(duplicate('Hello'))# 'HelloHello'
```
