```python
# using format for readability and efficiency
scores = [("Rodney Dangerfield", -1), ("Marlon Brando", 1), ("You", 100)] #list of tuples
for person in scores:
    name = person[0] # each tuple has indexes that can be accessed
    score = person[1] # second element of tuple
    print("Hello {}. Your score is {}.".format(name, score))
```

```python
person = 'Daniel'
greeting = 'Hello, {}!'.format(person) # takes person variable and inserts it into the string at the '{}' location
print(greeting)
```

```python
origPrice = float(input('Enter the original price: $'))
discount = float(input('Enter discount percentage: '))
newPrice = (1 - discount/100)*origPrice
calculation = '${} discounted by {}% is ${}.'.format(origPrice, discount, newPrice)
print(calculation)
```

```python
body_weight = 265
goal = 225
weight_loss = body_weight - goal

summary = 'You current weight is {} and you want to get to {} pounds. You will need to lose {} pounds.'.format(body_weight, goal, weight_loss)
print(summary)
```
