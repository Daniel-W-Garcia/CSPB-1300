```python
pirate = ['matey', 'fleabag inn', 'swabbie', 'matey', 'proud beauty',
'foul blaggart', 'galley', 'yer', 'arr', 'swabbies', 'be',
'foul blaggart', 'th’', 'head', 'me', 'avast', 'be', 'matey']

English = ['sir', 'hotel', 'student', 'boy', 'madam', 'professor', 'restaurant',
'your', 'excuse', 'students', 'are', 'lawyer', 'the', 'restroom',
'my', 'hello', 'is', 'man']

pirate_translations = {} # empty dict to hold eng/pir pairs
i = 0 # used to iterate and index pirate list
import string

# build dict from English and Pirate lists
for word in English:
    pirate_translations[word] = pirate_translations.get(word, pirate[i]) # add each Eng word as key and Pirate word as value
    i += 1 # iterate to match next loop of English list

punc = string.punctuation
translate = input('Please enter any phrase to be translated')

#grab end of sentence to add on at finish
end_sentence = ''
if translate[-1] in punc:
    end_sentence = translate[-1]
#strip out any punctuation
for c in translate:
    if c in punc:
        translate = translate.replace(c, '')

phrase = translate.split() #split phrase into words for comparison
eng_keys = list(pirate_translations.keys()) #cast keys into str

#sub in pirate words for English words
for word in phrase:
    if word in eng_keys:
        j = phrase.index(word)
        p_word = pirate_translations[word]
        phrase.remove(word)
        phrase.insert(j, p_word)
        output = " ".join(phrase)#turn new phrase list into a sentence (minus and punctuation)

print(output + end_sentence)
'''this seems to work, and handle the edge cases I could com put with
like '???' at the end of a sentence. Now that I solved, I will try to
optimize later'''
```
