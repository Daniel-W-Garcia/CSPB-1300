```python
socks = {'while': 5, 'black': 7, 'green': 14}

print(socks.get('black')) # the 'get' works like the index function but does not cause an error if the key isn't present
print(socks.get('brown')) # if the key is not present, 'get' will return 'None'
print(socks.get('brown', "no brown socks")) # overload allows to specify the return value if the key is not present
```

```python
with open('scarlet.txt', 'r') as file:
    text = file.read()
    letter_counts = {}

    for c in text: # every letter in the document
        if c not in letter_counts:
            letter_counts[c] = 0 # add the character as a key and initial value of 0
        letter_counts[c] += 1 # add 1 to character total. This catches the new values in the dict and all previous
print('t: ' + str(letter_counts['t'])+ ' occurrences')
```

```python
with open('scarlet.txt') as file:
    text = file.read()
    letters = {}
    for c in text:
        letters[c] = letters.get(c, 0) + 1 # this is really slick. using the 'get' keyword to check if c is in the dict
                                           # if it isn't '0' is custom return value and +1 either way
print('y ' + str(letters['y']) + ' occurrences')
```

```python
d = {'a': 194, 'b': 54, 'c':34, 'd': 44, 'e': 312, 'full':31}

ks = d.keys()
best_key_so_far = list(d.keys())[0]
for k in ks:
    if d[k] > d[best_key_so_far]:
        best_key_so_far = k

print("key " + best_key_so_far + " has the highest value, " + str(d[best_key_so_far]))
```

```python
placement = "Beaches are cool places to visit in spring however the Mackinaw Bridge is near. Most people visit Mackinaw later since the island is a cool place to explore."

d = {}
for c in placement:
    d[c] = d.get(c, 0) + 1 #get c or add it with val of 0. + 1 for all c


dk = list(d.keys()) # get list of keys
min_value = dk[0] # initialize min_value as first key in

for k in dk: # loop every hey's value
    if d[k] < d[min_value]: #compare with value at d[min_value]
        min_value = k # reassign with new min value
```
