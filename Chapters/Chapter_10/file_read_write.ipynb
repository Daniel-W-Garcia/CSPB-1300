```python
with open('math_file.txt', 'w') as file:
    for i in range(10):
        square = i*i
        file.write(str(square)+'\n') # can only write strings to files
```

```python
with open('math_file.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
with open('math_file.csv', 'w') as file:
    for i in range(10):
        square = i * i
        file.write(str(i)+','+str(square)+'\n')
```

```python
with open('math_file.csv', 'r') as file:
    content = file.read()
    print(content)
```
