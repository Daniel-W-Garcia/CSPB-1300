```python
import datetime

mean_SP = 0
max_interest = 0
count = 0
start_date = datetime.date(2016, 6, 1)
end_date = datetime.date(2017, 5, 1)

with open('SP500.txt', 'r') as file:
    file.readline()
    for line in file:
        line_parts = line.strip().split(',')
        date_parts = line_parts[0].split('/')
        m = int(date_parts[0])
        d = int(date_parts[1])
        y = int(date_parts[2])
        entry_date = datetime.date(y,m,d)
        if  start_date <= entry_date <= end_date:
            sp = float(line_parts[1])
            interest_rate = float(line_parts[5])
            mean_SP += sp
            count += 1
            if interest_rate > max_interest:
                max_interest = float(interest_rate)

mean_SP = mean_SP/count
print('{} is the mean SP and {} is the max interest in the given period.'.format(mean_SP, max_interest))
# OMG I know I took the longest and hardest possible route to solve
# this. I couldn't think of a way to get entries in a date range as
# the dates were str. This took me way longer than it should have

# one third = 1/3 = 1 divided by 3 which is .66 repeating
```
