```python
# important note: if you create a list then assign that list a new name, it becomes aliased. modifying either list will affect the other
a = [1,2,3]
b = a

b[3:] = [4,5,6]
print(b)
print(a) # eventhough we modified b it affects a since we did b = a. That makes them an alias and both b and a point to the same obj in memory
```

```python
#if we want an identical copy of a list and not an alias, it's easiest to use slice
c = ['Mike', 'John', 'Sally']
d = c[:]
print(d)
print(c)
d.append('Bob')
print(c)
print(d)# now we dont' affect the original
```
