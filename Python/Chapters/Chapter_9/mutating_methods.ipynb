```python
# some of the methods like append, sort, and reverse all return NONE. So if we try to increment a list with the function result
# it does not behave as such

my_list = []
my_list.append(1)
my_list.append(2)
my_list.append(3)
my_list.append(4)
print(my_list)
my_list = my_list.reverse() # this is not what we want. it cleared out all data and returned none
print(my_list)
```

```python
my_list_1 = []
my_list_1.append(1)
my_list_1.append(2)
my_list_1.append(3)
my_list_1.append(4)

print(my_list_1)
my_list_1.reverse() # modifies the object but doesn't create a new object.
print(my_list_1)
```

```python
# any method that is a mutator only will follow this same pattern.

j = []
j.append(1)
j.append(2)
j.append(3)
j.append(4)

i = reversed(j) # this makes an iterable object
print(i)
a = list(i) #can cast iterable to list
print(a)
k = list(reversed(j)) # cast to a list to return a reversed copy of original list in one line
print(k)
```

```python
# using the built in 'id' method can tell you if two variables point to the same object
b = []
print('id:', id(b))
b[:] = [10,20,30,40,50]
print('id:', id(b)) # append modifies the object and therefore id is the same
b.append(60)
print('id:', id(b))
b = b + [70]
print(b)
print('id:', id(b)) # b is a new obj with a new id after concatenation
```
