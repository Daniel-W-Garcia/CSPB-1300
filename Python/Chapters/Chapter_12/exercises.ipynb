```python
import random
def get_highest_random():
    highest = 0
    count = 0
    for _ in range(101):
        count += 1
        r_int = random.randrange(1, 1000)
        print('int: ', r_int)
        if r_int > highest:
            highest = r_int
            print('Highest: ', highest)
    print('Count: ', count)
    return highest

i = get_highest_random()

print(i)
```

```python
import math
def is_rightangled(a, b, c):
    a_acos = math.acos((b*b + c*c - a*a)/(2*b*c))
    a_angle = math.degrees(a_acos)
    print('A: ', a_angle)

    b_acos = math.acos((a*a + c*c - b*b)/(2*a*c))
    b_angle = math.degrees(b_acos)
    print('B: ', b_angle)

    c_angle = 180 - a_angle - b_angle
    print('C: ', c_angle)

    if abs(90 - c_angle) <.001:
        return True
    else:
        return False

is_rightangled(1.5,2.0,2.5)
```
