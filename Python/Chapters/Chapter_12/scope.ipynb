```python
'''
Local < Global < Built-In
As lines of code are interpreted, pythong looks up variable or methods in the local then global then built-in name spaces'
The first instance found is used.
'''

not_local = 5

def scope_example():
    local_var = 1 #this is a local variable
    this_var = not_local + 5 #this line gets a global var and executes an expression before assigning that value to  a local var
    sum_var = local_var + this_var
    print(local_var, '+', this_var, '=', sum_var)# 'print' is in the built-in scope and the last scope the interpreter will look for it
    return sum_var

print(scope_example())
```

```python
# Functions can include other functions

def adding(n: int) -> int:
    y = n + 3
    return y

#create list of n with a length of n + 3 (from adding func above)
def make_list(n: int) -> list:
    list_range = adding(n) #taking input, adding 3 and assigning as int to list_range
    created_list = []
    for i in range(list_range): # iterate n times based on adding method
        created_list.append(n) # add the input value to the list n + 3 times
    return created_list

cool_list = make_list(4) # assign the created list to a new, local var
print(cool_list)
```
