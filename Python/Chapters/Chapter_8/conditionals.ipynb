```python
'''
check if triangle is right angled
using law of cosines
once we have cos of an angle, we can use inverse cosine to get radians
convert radians to degrees
'''

import math
a = 5
b = 6
c = 8
is_rightangled = True

#cos(A) = (b^2 + c^2 - a^2)/(2bc)

cos_a = (b*b + c*c - a*a)/(2*b*c)
acos_a = math.acos(cos_a)
degrees_a = (acos_a * 180)/math.pi

cos_b = (a*a + c*c - b*b)/(2*a*c)
acos_b = math.acos(cos_b)
degrees_b = (acos_b * 180)/math.pi

degrees_c = 180 - (degrees_a + degrees_b)

if abs(180 - degrees_a + degrees_b + degrees_c) < .001:
    is_rightangled = True
else:
    is_rightangled = False


print(is_rightangled)

```

```python
#Reversing a string
my_str = 'abcd'
r = my_str[::-1] #this is a slicing operation. gets entire list starting at the back : start : stop : step
print(r)

r_str = ''
index = len(my_str) - 1 #start at the end of the string and work back

while index >= 0:
    r_str += my_str[index] #add each letter to the string
    index -= 1 #move backward one
print(r_str)

rr = ''
for char in my_str:
    rr = char + rr #this is kind of tricky. it's each letter in the front of rr. first iteration 1 rr = a + '' second iteration rr = b + a

print(rr)
```

```python
sentence = "students flock to the arb for a variety of outdoor activities such as jogging and picnicking"
sentence = sentence.split(' ')

same_letter_count = 0
first_letter = []
for word in sentence:
    first_letter.append(word[0])
for l in first_letter:
    count = first_letter.count(l)
    if count > 1:
        same_letter_count += 1

print(same_letter_count)


```
