```python
from fpdf import FPDF

pdf = FPDF()
pdf.set_auto_page_break(auto=True, margin=15)
pdf.add_page()
pdf.set_font("Courier", size=10)

with open("/home/daniel/Documents/Lab_1_Algorithm_and_Pseudocode.pdf.txt", "r", encoding="utf-8") as file:
    text = file.read()
    pdf.multi_cell(0, 6, text)

pdf.output('Lab_1.pdf', 'F')
```
