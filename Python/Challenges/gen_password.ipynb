```python

import random

letters = "abcdefghijklmnopqrstuvwxyz"
caps = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
numbers = "1234567890"

pw = ''
# Make an 8 letter password by combining characters from the three strings
for n in range(8):
    i = random.randrange(3)
    print('i:', i)
    if i == 0:
        char = random.choice(letters)
        print('Letters char:',random.choice(letters))
        pw += char
        print('Letters pw:',pw)
    elif i == 1:
        char = random.choice(caps)
        print('CAPS char:', random.choice(caps))
        pw += char
        print('CAPS pw:', char)
    else:
        char = random.choice(numbers)
        print('numbers char:', random.choice(numbers))
        pw += char
        print('numbers pw:',char)

print(pw)
```

```python
from itertools import product
from string import ascii_lowercase

sys.setExecutionLimit(60000) # 60 seconds
my_password = "abcd"
guess_num = 0
done = False

def brute_force(password):
    for guess_tuple in product(ascii_lowercase, repeat=4):
        guess = ''.join(guess_tuple)
        if guess == password:
            return guess
    return None
```
