```python
def calc_median(numbers: list) -> int:
    numbers.sort()
    val1 = int(len(numbers)/2) - 1
    val2 = int(len(numbers)/2)  #should be median in odd list

    if len(numbers) % 2 == 0:
        median = (numbers[val1] + numbers[val2])/2
    else:
        median = numbers[val2]
    return median

calc_median([1,2,3,4])
```
