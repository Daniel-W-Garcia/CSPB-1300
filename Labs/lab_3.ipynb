```python
import random

'''
need 2 pointers. if right pointer is smaller than left pointer, swap them
if not move right pointer 1 and check again. loop until the smallest number is in the left
move left pointer + 1
move right pointer  to L pointer +1 and loop as before
'''
range_to_sort = input('Please enter a number for how many integers we will sort')
range_to_sort = int(range_to_sort)
print(range_to_sort)

range_list = []

for num in range(range_to_sort):
    n = random.randrange(100)
    range_list.append(n)

print('Here are', range_to_sort, 'random numbers between 0-99', range_list)

l_pointer = 0
r_pointer = 0

while l_pointer < range_to_sort - 2:
    if r_pointer > 1:
        l_pointer += 1
        r_pointer = l_pointer + 1
    while r_pointer <= range_to_sort - 1:
        if range_list[r_pointer] < range_list[l_pointer]:
            temp = range_list[l_pointer]
            range_list[l_pointer] = range_list[r_pointer]
            range_list[r_pointer] = temp
            r_pointer += 1
        else:
            r_pointer += 1


print(range_list)
```
