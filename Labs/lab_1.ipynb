```python
invested = input('how much did you invest?')
invested_int = int(invested)

annual_loss_rate = input ('what is the annual loss rate percentage?')
annual_loss_rate = int(annual_loss_rate)
annual_loss_rate = annual_loss_rate/100
daily_loss_rate = annual_loss_rate/365

month_withdraw_amount = input('how much will you withdraw per month?')
month_withdraw_amount_int = int(month_withdraw_amount)

daily_loss_amount = invested_int * daily_loss_rate
current_balance = invested_int

total_withdrawn = 0
month_withdraws = 0
lost_half = False
days = 0
total_amount_lost = 0
net_balance = 0

while not lost_half:
    days += 1
    current_balance -= daily_loss_amount
    total_amount_lost += daily_loss_amount

    if current_balance < (invested_int - total_withdrawn) /2:
        lost_half = True
        print(lost_half)

    if days % 30 == 0:
        total_withdrawn += month_withdraw_amount_int
        month_withdraws += 1

get_days = days % 30
net_balance = current_balance + total_withdrawn

print('Current balance:',current_balance)
print('Total withdrawn:',total_withdrawn)
print('Total loss:',total_amount_lost)
print('Net balance:',net_balance)
print('You have lost half your invested money after', month_withdraws, 'months and ', get_days, 'days')
```
